# 1.Research Motivation

Modern machine learning models are widely used for high-stakes decision-making; however, their performance is often evaluated primarily using accuracy metrics, without sufficient consideration of fairness, interpretability, and bias. The Titanic dataset provides a controlled historical scenario to analyze how socio-demographic features influence survival predictions. This study is motivated by the need to move beyond predictive performance and examine how machine learning models learn patterns that may reflect underlying social biases. Understanding these behaviors is crucial for developing responsible and human-centric AI systems.

# 2.Research Gap
Most existing Titanic survival analyses focus on improving prediction accuracy through feature engineering or complex models, while largely ignoring interpretability and fairness analysis. Limited attention has been given to understanding how models behave across sensitive attributes such as gender and passenger class. Additionally, few studies combine model comparison with post-hoc interpretability and subgroup performance evaluation in a single pipeline. This gap motivates a comprehensive, research-oriented analysis that integrates prediction, explanation, and ethical evaluation.

# 3.Research objective
The primary objective of this research is to develop a structured machine learning pipeline that emphasizes not only predictive performance but also interpretability and fairness. The study aims to compare multiple classification models, analyze engineered features, and evaluate how predictions differ across demographic subgroups. By incorporating interpretability techniques, the research seeks to understand model decision mechanisms and potential biases. Ultimately, the goal is to promote transparent and responsible machine learning practices.

# 4.Research Question
- How do different machine learning models compare in predicting survival outcomes on the Titanic dataset?
- Which engineered features contribute most significantly to model decisions?
- Do trained models exhibit performance disparities across gender and passenger class?
- How can interpretability techniques help explain and evaluate these disparities?
